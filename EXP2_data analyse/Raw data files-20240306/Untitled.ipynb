```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
Materials=pd.read_excel('modulus.xlsx',sheet_name="Materials")
Polymers=pd.read_excel('modulus.xlsx',sheet_name="Polymers")
```

```python
def bin(x):
    bin=int(1+3.3*np.log(np.size(x)))
    return bin
```

```python
hist1,bin_edge1=np.histogram(Materials['modulus(MPa)'],bins=bin(Materials['modulus(MPa)']),
                            range=(np.min(Materials['modulus(MPa)']),np.max(Materials['modulus(MPa)'])))
fig1,ax1=plt.subplots()
ax1_twin=ax1.twinx()
ax1.hist(bin_edge1[:-1],bin_edge1,weights=hist1,facecolor='blue')
ax1.set_xlabel('Modulus(MPa)')
ax1.set_ylabel('Frequncy')
ax1_twin.set_ylabel('Probability')
mu=Materials['modulus(MPa)'].mean()
sigma=Materials['modulus(MPa)'].std()
x=np.linspace(Materials['modulus(MPa)'].min(),Materials['modulus(MPa)'].max(),100)
ndf=np.exp(-(x-mu)**2/(2*sigma**2))/(sigma*np.sqrt(2*np.pi))
ax1_twin.plot(x,ndf,'r-',label="PDF")
ax1_twin.legend(loc='upper right')
plt.tight_layout()
plt.grid(True)
plt.title("The distribution of Material's group")
plt.show()
```

```python
stats.probplot(Materials['modulus(MPa)'], dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid(True)
plt.show()
```
